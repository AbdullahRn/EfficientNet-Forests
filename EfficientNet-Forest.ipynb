```python
# ================================================================
# CELL 1: SETUP — Kaggle Notebook Version
# ================================================================
!pip install -q ptflops

import os, json, random, time, copy
from pathlib import Path

import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset
from torchvision import datasets, transforms, models
from sklearn.model_selection import train_test_split
import torch.nn.utils.prune as prune
import matplotlib.pyplot as plt
from ptflops import get_model_complexity_info

# -------------------------
#  Dataset path (Kaggle)
# -------------------------

LOCAL_VEHICLES_DIR = "/kaggle/input/datasets/naifislam/paribahan-bd/Local-Vehicles/Local-Vehicles"

print("Dataset exists:", os.path.exists(LOCAL_VEHICLES_DIR))
print("Vehicle class folders:", os.listdir(LOCAL_VEHICLES_DIR))

# -------------------------
#  Reproducibility
# -------------------------
GLOBAL_SEED = 42
random.seed(GLOBAL_SEED)
np.random.seed(GLOBAL_SEED)
torch.manual_seed(GLOBAL_SEED)
torch.cuda.manual_seed_all(GLOBAL_SEED)

torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

# -------------------------
#  Device
# -------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
# ================================================================
# CELL 2: DATA LOADING + SPLIT + TRANSFORMS
# ================================================================
IMG_SIZE = 224
BATCH_SIZE = 32
NUM_WORKERS = 2  # Kaggle supports >0

# -------------------------
#  Base dataset (no transform)
# -------------------------
base_dataset = datasets.ImageFolder(root=LOCAL_VEHICLES_DIR, transform=None)
print("Found classes:", base_dataset.classes)
NUM_CLASSES = len(base_dataset.classes)

indices = np.arange(len(base_dataset))
labels = np.array(base_dataset.targets)

train_idx, tmp_idx, y_train, y_tmp = train_test_split(
    indices, labels, test_size=0.3, random_state=GLOBAL_SEED, stratify=labels
)
val_idx, test_idx, y_val, y_test = train_test_split(
    tmp_idx, y_tmp, test_size=0.5, random_state=GLOBAL_SEED, stratify=y_tmp
)

print(f"TOTAL SAMPLES: {len(base_dataset)}")
print(f"TRAIN: {len(train_idx)} | VAL: {len(val_idx)} | TEST: {len(test_idx)}")

# -------------------------
#  Transforms
# -------------------------
train_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.RandomResizedCrop(IMG_SIZE, scale=(0.8, 1.0)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225]),
])

eval_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.CenterCrop(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225]),
])

# -------------------------
#  Custom Dataset
# -------------------------
class IndexedImageFolder(Dataset):
    def __init__(self, base_ds, indices, transform=None):
        self.base = base_ds
        self.indices = list(indices)
        self.transform = transform

    def __len__(self):
        return len(self.indices)

    def __getitem__(self, i):
        img, label = self.base[self.indices[i]]
        img = img.convert("RGB")
        if self.transform:
            img = self.transform(img)
        return img, label

train_dataset = IndexedImageFolder(base_dataset, train_idx, transform=train_transform)
val_dataset   = IndexedImageFolder(base_dataset, val_idx,   transform=eval_transform)
test_dataset  = IndexedImageFolder(base_dataset, test_idx,  transform=eval_transform)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE,
                          shuffle=True, num_workers=NUM_WORKERS, pin_memory=True)
val_loader   = DataLoader(val_dataset, batch_size=BATCH_SIZE,
                          shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)
test_loader  = DataLoader(test_dataset, batch_size=BATCH_SIZE,
                          shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)
```

```python
# ================================================================
# CELL 3: CONFIG + UTILS
# ================================================================
EXPERT_BACKBONES = ["efficientnet_b0", "efficientnet_b1", "efficientnet_b2"]
NUM_EXPERTS = len(EXPERT_BACKBONES)
BOOTSTRAP_RATIO = 0.7

TRIAL_EPOCHS = 1
FULL_EPOCHS  = 2

LAMBDA_GL = 1e-5
PRUNE_RATIOS = [0.05, 0.10, 0.15]
PRUNE_FT_EPOCHS = 2

KD_TEMPERATURE = 4.0
KD_ALPHA = 0.3
KD_EPOCHS = 3

LR_EXPERT = 1e-4
LR_STUDENT = 1e-3

criterion_ce = nn.CrossEntropyLoss()

# -------------------------
# Utilities
# -------------------------
def count_params(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

def group_lasso_penalty(model):
    gl = 0.0
    for m in model.modules():
        if isinstance(m, nn.Conv2d):
            w = m.weight
            w_norm = w.pow(2).sum(dim=(1,2,3)).sqrt()
            gl += w_norm.sum()
    return gl

def compute_flops(model, img_size=224):
    model_cpu = copy.deepcopy(model).to("cpu")
    model_cpu.eval()
    with torch.no_grad():
        macs, params = get_model_complexity_info(
            model_cpu, (3, img_size, img_size),
            as_strings=False,
            print_per_layer_stat=False
        )
    return macs/1e9, params/1e6  # GMACs, MParams

def measure_latency(model, device, runs=100, warmup=20):
    model = model.to(device)
    model.eval()
    dummy = torch.randn(1, 3, IMG_SIZE, IMG_SIZE).to(device)

    with torch.no_grad():
        for _ in range(warmup):
            _ = model(dummy)

    if device.type == "cuda":
        torch.cuda.synchronize()

    start = time.time()
    with torch.no_grad():
        for _ in range(runs):
            _ = model(dummy)
    if device.type == "cuda":
        torch.cuda.synchronize()
    end = time.time()

    return (end - start) / runs * 1000.0

def measure_throughput(model, device, batch_size=32, runs=50, warmup=10):
    model = model.to(device)
    model.eval()
    dummy = torch.randn(batch_size, 3, IMG_SIZE, IMG_SIZE).to(device)

    with torch.no_grad():
        for _ in range(warmup):
            _ = model(dummy)

    if device.type == "cuda":
        torch.cuda.synchronize()

    start = time.time()
    with torch.no_grad():
        for _ in range(runs):
            _ = model(dummy)
    if device.type == "cuda":
        torch.cuda.synchronize()
    end = time.time()

    total_images = runs * batch_size
    return total_images / (end - start)  # images/sec

def full_test_inference_time(model, dataloader, device):
    model = model.to(device)
    model.eval()

    if device.type == "cuda":
        torch.cuda.synchronize()

    start = time.time()
    with torch.no_grad():
        for images, _ in dataloader:
            images = images.to(device)
            _ = model(images)
    if device.type == "cuda":
        torch.cuda.synchronize()
    end = time.time()

    total_images = len(dataloader.dataset)
    total_time = end - start
    return total_time, (total_time/total_images)*1000.0

def model_size_mb(model, filename="temp_model.pth"):
    torch.save(model.state_dict(), filename)
    size_mb = os.path.getsize(filename) / (1024*1024)
    os.remove(filename)
    return size_mb
```

```python
# ================================================================
# CELL 4: MODELS + TRAIN/EVAL LOOPS
# ================================================================
def build_efficientnet_expert(name, num_classes):
    if name == "efficientnet_b0":
        m = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1)
    elif name == "efficientnet_b1":
        m = models.efficientnet_b1(weights=models.EfficientNet_B1_Weights.IMAGENET1K_V1)
    elif name == "efficientnet_b2":
        m = models.efficientnet_b2(weights=models.EfficientNet_B2_Weights.IMAGENET1K_V1)
    else:
        raise ValueError("Unknown EfficientNet variant")

    in_features = m.classifier[-1].in_features
    m.classifier[-1] = nn.Linear(in_features, num_classes)
    return m.to(device)

def build_student_model(num_classes):
    m = models.mobilenet_v3_small(weights=models.MobileNet_V3_Small_Weights.IMAGENET1K_V1)
    in_features = m.classifier[-1].in_features
    m.classifier[-1] = nn.Linear(in_features, num_classes)
    return m.to(device)

def train_one_epoch(model, dataloader, optimizer, criterion, device, lambda_gl=0.0):
    model.train()
    total_loss, total_correct, total = 0.0, 0, 0

    for images, labels in dataloader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        out = model(images)
        ce_loss = criterion(out, labels)

        loss = ce_loss + lambda_gl * group_lasso_penalty(model)
        loss.backward()
        optimizer.step()

        total_loss += ce_loss.item() * images.size(0)
        total_correct += (out.argmax(1) == labels).sum().item()
        total += labels.size(0)

    return total_loss/total, total_correct/total

def evaluate(model, dataloader, criterion, device):
    model.eval()
    total_loss, total_correct, total = 0.0, 0, 0
    with torch.no_grad():
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)
            out = model(images)
            loss = criterion(out, labels)

            total_loss += loss.item() * images.size(0)
            total_correct += (out.argmax(1) == labels).sum().item()
            total += labels.size(0)

    return total_loss/total, total_correct/total
```

```python
# ================================================================
# CELL 5: TRAIN EXPERTS
# ================================================================
experts = []
for name in EXPERT_BACKBONES:
    model = build_efficientnet_expert(name, NUM_CLASSES)
    print(f"Built expert {name} with {count_params(model)/1e6:.2f}M params")
    experts.append(model)

expert_val_scores = []

for i, expert in enumerate(experts):
    print(f"\n=== Training Expert {i+1}/{NUM_EXPERTS} ===")

    n_train = len(train_dataset)
    sub_size = int(BOOTSTRAP_RATIO * n_train)
    sub_idx = np.random.choice(np.arange(n_train), size=sub_size, replace=True)
    train_subset = torch.utils.data.Subset(train_dataset, sub_idx)

    train_loader_expert = DataLoader(train_subset, batch_size=BATCH_SIZE,
                                     shuffle=True, num_workers=NUM_WORKERS, pin_memory=True)

    optimizer = torch.optim.AdamW(expert.parameters(), lr=LR_EXPERT, weight_decay=1e-4)

    # Trial
    for epoch in range(TRIAL_EPOCHS):
        t_loss, t_acc = train_one_epoch(expert, train_loader_expert, optimizer, criterion_ce, device, LAMBDA_GL)
        v_loss, v_acc = evaluate(expert, val_loader, criterion_ce, device)
        print(f"[Trial {epoch+1}] Train Acc={t_acc:.3f}, Val Acc={v_acc:.3f}")

    # Full
    for epoch in range(FULL_EPOCHS):
        t_loss, t_acc = train_one_epoch(expert, train_loader_expert, optimizer, criterion_ce, device, LAMBDA_GL)
        v_loss, v_acc = evaluate(expert, val_loader, criterion_ce, device)
        print(f"[Full {epoch+1}] Train Acc={t_acc:.3f}, Val Acc={v_acc:.3f}")

    expert_val_scores.append(v_acc)

print("\nValidation scores:", expert_val_scores)
```

```python
# ================================================================
# CELL 6: STRUCTURED PRUNING
# ================================================================
def structured_channel_prune(model, amount):
    for m in model.modules():
        if isinstance(m, nn.Conv2d):
            prune.ln_structured(m, name="weight", amount=amount, n=2, dim=0)
            prune.remove(m, "weight")

for i, expert in enumerate(experts):
    print(f"\n=== Pruning Expert {i+1} (ratio search) ===")

    base_state = copy.deepcopy(expert.state_dict())
    pre_val = expert_val_scores[i]
    best_val = pre_val
    best_ratio = 0.0
    best_state = base_state

    for ratio in PRUNE_RATIOS:
        print(f"\n  -> Trying prune ratio = {ratio:.2f}")
        temp_model = copy.deepcopy(expert)
        structured_channel_prune(temp_model, ratio)

        optimizer = torch.optim.AdamW(temp_model.parameters(), lr=LR_EXPERT)

        for ft_epoch in range(PRUNE_FT_EPOCHS):
            t_loss, t_acc = train_one_epoch(temp_model, train_loader, optimizer, criterion_ce, device, lambda_gl=0.0)

        v_loss, v_acc = evaluate(temp_model, val_loader, criterion_ce, device)
        print(f"     Val Acc after FT: {v_acc:.3f}")

        if v_acc >= pre_val * 0.95 and v_acc > best_val:
            best_val = v_acc
            best_ratio = ratio
            best_state = copy.deepcopy(temp_model.state_dict())

    if best_ratio > 0.0:
        print(f"\nChosen prune ratio: {best_ratio:.2f}, Val={best_val:.3f}")
        expert.load_state_dict(best_state)
        expert_val_scores[i] = best_val
    else:
        print("\nNo pruning accepted; keeping original.")
        expert.load_state_dict(base_state)
```

```python
# ================================================================
# CELL 7: FOREST + STUDENT KD
# ================================================================
val_scores = np.array(expert_val_scores)
forest_weights_np = val_scores / val_scores.sum()
forest_weights = torch.tensor(forest_weights_np, dtype=torch.float32, device=device)
print("Forest Weights:", forest_weights.cpu().tolist())

def forest_logits(experts, images, weights):
    for m in experts:
        m.eval()
    with torch.no_grad():
        logits = torch.stack([m(images) for m in experts])  # [E,B,C]
    return (logits * weights.view(-1,1,1)).sum(0)

student = build_student_model(NUM_CLASSES)
print("Student params:", count_params(student)/1e6, "M")

def kd_loss_fn(s_logits, t_logits, labels, T, alpha):
    ce = criterion_ce(s_logits, labels)
    log_p_s = F.log_softmax(s_logits / T, dim=1)
    p_t = F.softmax(t_logits / T, dim=1)
    kd = F.kl_div(log_p_s, p_t, reduction="batchmean") * (T**2)
    return alpha*ce + (1-alpha)*kd

optimizer_student = torch.optim.AdamW(student.parameters(), lr=LR_STUDENT)

print("\n=== TRAINING STUDENT WITH KD ===")
for epoch in range(KD_EPOCHS):
    student.train()
    total_correct, total = 0, 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        with torch.no_grad():
            t_logits = forest_logits(experts, images, forest_weights)

        optimizer_student.zero_grad()
        s_logits = student(images)
        loss = kd_loss_fn(s_logits, t_logits, labels, KD_TEMPERATURE, KD_ALPHA)

        loss.backward()
        optimizer_student.step()

        total_correct += (s_logits.argmax(1) == labels).sum().item()
        total += labels.size(0)

    train_acc = total_correct / total
    val_acc = evaluate(student, val_loader, criterion_ce, device)[1]
    print(f"[KD {epoch+1}] Train Acc={train_acc:.3f}, Val Acc={val_acc:.3f}")

test_acc = evaluate(student, test_loader, criterion_ce, device)[1]
print("Final Student Test Accuracy:", test_acc)
```

```python
# ================================================================
# CELL 8: FULL METRICS (Accuracy + Params + FLOPs + Inference Time)
# ================================================================
rows = []

# helper to get accuracy
def eval_top1(model, loader):
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for x,y in loader:
            x,y = x.to(device), y.to(device)
            out = model(x)
            correct += (out.argmax(1)==y).sum().item()
            total += y.size(0)
    return correct/total

# NEW: Helper function to evaluate forest ensemble
def eval_forest_top1(experts, loader, weights):
    # Move all experts to the correct device (should be CUDA)
    for expert in experts:
        expert.to(device)
        expert.eval()

    correct, total = 0, 0
    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)
            logits = torch.stack([expert(images) for expert in experts])  # [E,B,C]
            ensemble_logits = (logits * weights.view(-1,1,1)).sum(0)
            correct += (ensemble_logits.argmax(1) == labels).sum().item()
            total += labels.size(0)
    return correct/total

# Calculate forest accuracy
# Ensure all experts are on CUDA
for expert in experts:
    expert.to(device)

forest_val_acc = eval_forest_top1(experts, val_loader, forest_weights)
forest_test_acc = eval_forest_top1(experts, test_loader, forest_weights)

print(f"Forest Ensemble Validation Accuracy: {forest_val_acc:.4f}")
print(f"Forest Ensemble Test Accuracy: {forest_test_acc:.4f}")

# -------------------------
# Forest Ensemble (added to benchmark)
# -------------------------
# Calculate average params and FLOPs for forest
forest_params_m = sum(count_params(expert) for expert in experts) / 1e6
forest_gmacs = sum(compute_flops(expert)[0] for expert in experts)

# Calculate latency for forest (approximate as max of experts since they run sequentially)
# Convert the comprehension to a loop to avoid scope issues
cpu_lat_list = []
for expert in experts:
    # Temporarily move expert to CPU for latency measurement
    expert_cpu = copy.deepcopy(expert).cpu()
    cpu_lat_list.append(measure_latency(expert_cpu, device=torch.device("cpu"), runs=20))
cpu_lat_forest = max(cpu_lat_list)

cpu_tp_list = []
for expert in experts:
    # Temporarily move expert to CPU for throughput measurement
    expert_cpu = copy.deepcopy(expert).cpu()
    cpu_tp_list.append(measure_throughput(expert_cpu, device=torch.device("cpu"), batch_size=32, runs=10))
cpu_tp_forest = min(cpu_tp_list)

# Calculate model size for forest
forest_model_mb = sum(model_size_mb(expert) for expert in experts)

# Calculate inference time for forest - FIXED VERSION
def forest_inference_time(experts, loader, weights, target_device):
    # Create device object from string
    dev = torch.device(target_device)

    # Move weights to the target device
    weights = weights.to(dev)

    # Move experts to target device
    experts_on_device = []
    for expert in experts:
        expert_on_device = copy.deepcopy(expert).to(dev)
        expert_on_device.eval()
        experts_on_device.append(expert_on_device)

    if dev.type == "cuda":
        torch.cuda.synchronize()

    start = time.time()
    with torch.no_grad():
        for images, _ in loader:
            images = images.to(dev)
            logits = torch.stack([expert(images) for expert in experts_on_device])
            _ = (logits * weights.view(-1,1,1)).sum(0)

    if dev.type == "cuda":
        torch.cuda.synchronize()
    end = time.time()

    total_images = len(loader.dataset)
    total_time = end - start
    return total_time, (total_time/total_images)*1000.0

# FIX: Move forest_weights to CPU first, then pass to function
forest_weights_cpu = forest_weights.cpu()
forest_total_t, forest_ms_img = forest_inference_time(experts, test_loader, forest_weights_cpu, "cpu")

# Add forest to rows
rows.append({
    "model": "Forest_Ensemble",
    "params_M": forest_params_m,
    "FLOPs_GMACs": forest_gmacs,
    "val_acc": forest_val_acc,
    "test_acc": forest_test_acc,
    "cpu_latency_ms": cpu_lat_forest,
    "cpu_throughput_img_s": cpu_tp_forest,
    "gpu_latency_ms": None,  # Forest runs experts sequentially, so latency would be sum
    "model_size_MB": forest_model_mb,
    "test_infer_total_s_CPU": forest_total_t,
    "test_infer_ms_per_img_CPU": forest_ms_img
})

# -------------------------
# Experts (individual)
# -------------------------
for i, (name, expert) in enumerate(zip(EXPERT_BACKBONES, experts), start=1):
    params_m = count_params(expert)/1e6
    gmacs, _ = compute_flops(expert)
    val_acc = eval_top1(expert, val_loader)
    test_acc = eval_top1(expert, test_loader)

    # For CPU measurements, use a CPU copy
    expert_cpu = copy.deepcopy(expert).cpu()
    cpu_lat = measure_latency(expert_cpu, device=torch.device("cpu"), runs=50)
    cpu_tp  = measure_throughput(expert_cpu, device=torch.device("cpu"), batch_size=32, runs=20)
    model_mb = model_size_mb(expert)

    # For full test inference, use CPU copy
    total_t, ms_img = full_test_inference_time(expert_cpu, test_loader, device=torch.device("cpu"))

    gpu_lat = None
    if device.type == "cuda":
        gpu_lat = measure_latency(expert, device=device, runs=50)

    rows.append({
        "model": f"Expert_{i} ({name})",
        "params_M": params_m,
        "FLOPs_GMACs": gmacs,
        "val_acc": val_acc,
        "test_acc": test_acc,
        "cpu_latency_ms": cpu_lat,
        "cpu_throughput_img_s": cpu_tp,
        "gpu_latency_ms": gpu_lat,
        "model_size_MB": model_mb,
        "test_infer_total_s_CPU": total_t,
        "test_infer_ms_per_img_CPU": ms_img
    })

# -------------------------
# Student
# -------------------------
params_m = count_params(student)/1e6
gmacs, _ = compute_flops(student)
val_acc = eval_top1(student, val_loader)
test_acc = eval_top1(student, test_loader)

# For CPU measurements, use a CPU copy
student_cpu = copy.deepcopy(student).cpu()
cpu_lat = measure_latency(student_cpu, device=torch.device("cpu"), runs=50)
cpu_tp  = measure_throughput(student_cpu, device=torch.device("cpu"), batch_size=32, runs=20)
model_mb = model_size_mb(student)

# For full test inference, use CPU copy
total_t, ms_img = full_test_inference_time(student_cpu, test_loader, device=torch.device("cpu"))

gpu_lat = None
if device.type == "cuda":
    gpu_lat = measure_latency(student, device=device, runs=50)

rows.append({
    "model": "Student_MobileNetV3_Small",
    "params_M": params_m,
    "FLOPs_GMACs": gmacs,
    "val_acc": val_acc,
    "test_acc": test_acc,
    "cpu_latency_ms": cpu_lat,
    "cpu_throughput_img_s": cpu_tp,
    "gpu_latency_ms": gpu_lat,
    "model_size_MB": model_mb,
    "test_infer_total_s_CPU": total_t,
    "test_infer_ms_per_img_CPU": ms_img
})

full_df = pd.DataFrame(rows)
print("\n=== FULL BENCHMARK TABLE ===")
print(full_df)

full_df.to_csv("full_benchmark_results.csv", index=False)
print("\nSaved: full_benchmark_results.csv")
```

```python
# ================================================================
# CELL 9: TEASER NUMBERS
# ================================================================
def teaser(df, baseline_model, student_model="Student_MobileNetV3_Small"):
    base = df[df["model"]==baseline_model].iloc[0]
    stud = df[df["model"]==student_model].iloc[0]

    params_red = 100*(1-stud["params_M"]/base["params_M"])
    flops_red  = 100*(1-stud["FLOPs_GMACs"]/base["FLOPs_GMACs"])
    acc_drop   = (stud["test_acc"]-base["test_acc"])*100

    print("\n=== TEASER ===")
    print("Baseline:", baseline_model)
    print("Student :", student_model)
    print(f"Params reduction: {params_red:.2f}%")
    print(f"FLOPs reduction : {flops_red:.2f}%")
    print(f"Accuracy drop   : {acc_drop:+.2f} pp")

# compare student vs best expert
best_expert = full_df.loc[full_df["test_acc"].idxmax()]["model"]
teaser(full_df, best_expert)
```

```python
# ================================================================
# CELL 10: SAVE ARTIFACTS + GENERATE PAPER PLOTS/TABLES (SAFE)
# Run this AFTER Cell 9
# ================================================================
# This cell does NOT retrain or modify the pipeline.
# It only:
#   - saves already-trained models
#   - saves benchmark tables
#   - computes predictions from the final saved-in-memory models
#   - generates the paper plots / confusion matrices / tables
# so your main values/claims remain unchanged.

import os
import json
import copy
from pathlib import Path

import numpy as np
import pandas as pd
import torch
import matplotlib.pyplot as plt

from sklearn.metrics import (
    precision_recall_fscore_support,
    balanced_accuracy_score,
    confusion_matrix,
)
from scipy.stats import chi2

# -------------------------
# Safety checks
# -------------------------
required_vars = [
    "experts", "student", "forest_weights",
    "val_loader", "test_loader", "device",
    "EXPERT_BACKBONES", "base_dataset", "full_df"
]
missing = [v for v in required_vars if v not in globals()]
if missing:
    raise RuntimeError(f"Missing required variables from earlier cells: {missing}")

# -------------------------
# Output folders
# -------------------------
ARTIFACT_ROOT = Path("/kaggle/working/saved_artifacts")
CKPT_DIR = ARTIFACT_ROOT / "checkpoints"
PLOTS_DIR = ARTIFACT_ROOT / "plots"
TABLES_DIR = ARTIFACT_ROOT / "tables"
PREDS_DIR = ARTIFACT_ROOT / "predictions"

for d in [ARTIFACT_ROOT, CKPT_DIR, PLOTS_DIR, TABLES_DIR, PREDS_DIR]:
    d.mkdir(parents=True, exist_ok=True)

class_names = list(base_dataset.classes)

# -------------------------
# Helper functions
# -------------------------
def eval_with_preds(model, loader, device):
    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for x, y in loader:
            x = x.to(device)
            out = model(x)
            preds = out.argmax(1).cpu().numpy()
            y_pred.extend(preds.tolist())
            y_true.extend(y.numpy().tolist())
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    acc = (y_true == y_pred).mean()
    return acc, y_true, y_pred

def forest_preds(experts, loader, weights, device):
    for expert in experts:
        expert.to(device)
        expert.eval()

    if not torch.is_tensor(weights):
        weights = torch.tensor(weights, dtype=torch.float32, device=device)
    else:
        weights = weights.to(device)

    y_true, y_pred = [], []
    with torch.no_grad():
        for images, labels in loader:
            images = images.to(device)
            logits = torch.stack([expert(images) for expert in experts])  # [E,B,C]
            ensemble_logits = (logits * weights.view(-1, 1, 1)).sum(0)
            preds = ensemble_logits.argmax(1).cpu().numpy()
            y_pred.extend(preds.tolist())
            y_true.extend(labels.numpy().tolist())

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    acc = (y_true == y_pred).mean()
    return acc, y_true, y_pred

def compute_macro_metrics(y_true, y_pred):
    prec, rec, f1, _ = precision_recall_fscore_support(
        y_true, y_pred, average="macro", zero_division=0
    )
    bal = balanced_accuracy_score(y_true, y_pred)
    return {
        "Macro Prec": prec * 100.0,
        "Macro Rec": rec * 100.0,
        "Macro F1": f1 * 100.0,
        "Balanced Acc": bal * 100.0,
    }

def mcnemar_test(y_true, pred_a, pred_b):
    a_correct = pred_a == y_true
    b_correct = pred_b == y_true

    b = int(np.sum((~a_correct) & (b_correct)))
    c = int(np.sum((a_correct) & (~b_correct)))

    if b + c == 0:
        chi2_stat = 0.0
        p_value = 1.0
    else:
        chi2_stat = ((abs(b - c) - 1) ** 2) / (b + c)
        p_value = 1 - chi2.cdf(chi2_stat, df=1)

    return {"b": b, "c": c, "chi2": chi2_stat, "p_value": p_value}

def save_confusion_plot(cm_norm, class_names, title, out_path):
    plt.figure(figsize=(8, 7))
    plt.imshow(cm_norm, interpolation="nearest")
    plt.title(title)
    plt.colorbar()
    ticks = np.arange(len(class_names))
    plt.xticks(ticks, class_names, rotation=90)
    plt.yticks(ticks, class_names)
    plt.tight_layout()
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.savefig(out_path, dpi=300, bbox_inches="tight")
    plt.close()
    print(f"Saved: {out_path}")

def save_confusion_combined_vertical(cm1, cm2, class_names, out_path):
    fig, axes = plt.subplots(2, 1, figsize=(10, 16))

    im1 = axes[0].imshow(cm1, interpolation="nearest")
    axes[0].set_title("Normalized Confusion Matrix - Forest Teacher")
    fig.colorbar(im1, ax=axes[0], fraction=0.046, pad=0.04)
    ticks = np.arange(len(class_names))
    axes[0].set_xticks(ticks)
    axes[0].set_xticklabels(class_names, rotation=90)
    axes[0].set_yticks(ticks)
    axes[0].set_yticklabels(class_names)
    axes[0].set_ylabel("True label")
    axes[0].set_xlabel("Predicted label")

    im2 = axes[1].imshow(cm2, interpolation="nearest")
    axes[1].set_title("Normalized Confusion Matrix - Student (Ours)")
    fig.colorbar(im2, ax=axes[1], fraction=0.046, pad=0.04)
    axes[1].set_xticks(ticks)
    axes[1].set_xticklabels(class_names, rotation=90)
    axes[1].set_yticks(ticks)
    axes[1].set_yticklabels(class_names)
    axes[1].set_ylabel("True label")
    axes[1].set_xlabel("Predicted label")

    plt.tight_layout()
    plt.savefig(out_path, dpi=300, bbox_inches="tight")
    plt.close()
    print(f"Saved: {out_path}")

def plot_bar(values, labels, ylabel, title, out_path):
    plt.figure(figsize=(8, 4.5))
    plt.bar(labels, values)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(rotation=20, ha="right")
    plt.tight_layout()
    plt.savefig(out_path, dpi=300, bbox_inches="tight")
    plt.close()
    print(f"Saved: {out_path}")

def plot_dual_metric(labels, gmacs, latency, out_path):
    fig, ax1 = plt.subplots(figsize=(8, 4.5))
    x = np.arange(len(labels))
    width = 0.35

    ax1.bar(x - width / 2, gmacs, width=width, label="GMACs")
    ax1.set_ylabel("GMACs")
    ax1.set_xticks(x)
    ax1.set_xticklabels(labels, rotation=20, ha="right")

    ax2 = ax1.twinx()
    ax2.bar(x + width / 2, latency, width=width, label="Latency (ms)")
    ax2.set_ylabel("Latency (ms)")

    plt.title("Compute Complexity and Latency")
    fig.tight_layout()
    plt.savefig(out_path, dpi=300, bbox_inches="tight")
    plt.close()
    print(f"Saved: {out_path}")

# -------------------------
# Save current trained models / weights
# -------------------------
for i, (name, expert) in enumerate(zip(EXPERT_BACKBONES, experts), start=1):
    torch.save(expert.state_dict(), CKPT_DIR / f"expert_{i}_{name}.pth")
torch.save(student.state_dict(), CKPT_DIR / "student_mobilenetv3_small.pth")

forest_weights_np = forest_weights.detach().cpu().numpy() if torch.is_tensor(forest_weights) else np.asarray(forest_weights)
np.save(CKPT_DIR / "forest_weights.npy", forest_weights_np)

with open(CKPT_DIR / "class_names.json", "w") as f:
    json.dump(class_names, f, indent=2)

print("Saved checkpoints and metadata.")

# -------------------------
# Save the exact benchmark table already produced by the notebook
# -------------------------
benchmark_df = full_df.copy()
benchmark_df.to_csv(ARTIFACT_ROOT / "full_benchmark_results.csv", index=False)
print("Saved:", ARTIFACT_ROOT / "full_benchmark_results.csv")

# -------------------------
# Predictions
# -------------------------
pred_map = {}
y_test_true = None

for i, (name, expert) in enumerate(zip(EXPERT_BACKBONES, experts), start=1):
    _, y_true_tmp, pred_tmp = eval_with_preds(expert, test_loader, device)
    pred_map[f"Expert {i} ({name.replace('efficientnet_', '').upper()})"] = pred_tmp
    if y_test_true is None:
        y_test_true = y_true_tmp

forest_test_acc, y_true_forest, forest_test_pred = forest_preds(experts, test_loader, forest_weights, device)
y_test_true = y_true_forest
pred_map["Forest Teacher"] = forest_test_pred

student_test_acc, _, student_test_pred = eval_with_preds(student, test_loader, device)
pred_map["Student (Ours)"] = student_test_pred

np.save(PREDS_DIR / "y_true.npy", y_test_true)
for key, arr in pred_map.items():
    safe_name = key.lower().replace(" ", "_").replace("(", "").replace(")", "").replace("/", "_")
    np.save(PREDS_DIR / f"{safe_name}.npy", arr)

# -------------------------
# Macro metrics
# -------------------------
macro_rows = []
for key, pred in pred_map.items():
    met = compute_macro_metrics(y_test_true, pred)
    macro_rows.append({
        "Model": key,
        "Macro Prec": met["Macro Prec"],
        "Macro Rec": met["Macro Rec"],
        "Macro F1": met["Macro F1"],
        "Balanced Acc": met["Balanced Acc"],
    })

macro_df = pd.DataFrame(macro_rows)
macro_df.to_csv(TABLES_DIR / "table_macro_metrics.csv", index=False)
print("Saved:", TABLES_DIR / "table_macro_metrics.csv")
print(macro_df)

# -------------------------
# Confusion matrices
# -------------------------
cm_forest = confusion_matrix(y_test_true, forest_test_pred, normalize="true")
cm_student = confusion_matrix(y_test_true, student_test_pred, normalize="true")

save_confusion_plot(
    cm_forest, class_names,
    "Normalized Confusion Matrix - Forest Teacher",
    PLOTS_DIR / "confusion_forest_blues.png"
)
save_confusion_plot(
    cm_student, class_names,
    "Normalized Confusion Matrix - Student",
    PLOTS_DIR / "confusion_student_blues.png"
)
save_confusion_combined_vertical(
    cm_forest, cm_student, class_names,
    PLOTS_DIR / "confusion_combined_vertical.png"
)

# -------------------------
# McNemar
# -------------------------
expert_only = benchmark_df[benchmark_df["model"].str.startswith("Expert_")].copy()
best_expert_row = expert_only.sort_values("test_acc", ascending=False).iloc[0]["model"]

if "efficientnet_b0" in best_expert_row:
    best_expert_pred = pred_map["Expert 1 (B0)"]
elif "efficientnet_b1" in best_expert_row:
    best_expert_pred = pred_map["Expert 2 (B1)"]
else:
    best_expert_pred = pred_map["Expert 3 (B2)"]

mcn_best = mcnemar_test(y_test_true, forest_test_pred, best_expert_pred)
mcn_student = mcnemar_test(y_test_true, forest_test_pred, student_test_pred)

mcn_df = pd.DataFrame([
    {
        "Comparison": f"Forest vs {best_expert_row}",
        "b": mcn_best["b"],
        "c": mcn_best["c"],
        "chi2": mcn_best["chi2"],
        "p_value": mcn_best["p_value"],
    },
    {
        "Comparison": "Forest vs Student (Ours)",
        "b": mcn_student["b"],
        "c": mcn_student["c"],
        "chi2": mcn_student["chi2"],
        "p_value": mcn_student["p_value"],
    },
])
mcn_df.to_csv(TABLES_DIR / "table_mcnemar.csv", index=False)
print("Saved:", TABLES_DIR / "table_mcnemar.csv")
print(mcn_df)

# -------------------------
# Build the paper-style table using the notebook's exact benchmark values
# -------------------------
def get_row(df, model_name):
    return df[df["model"] == model_name].iloc[0]

def pick_latency_ms(row, is_forest=False):
    # Prefer the exact field already computed in the notebook.
    # For forest, the notebook stored the paper-useful serial test ms/image in test_infer_ms_per_img_CPU.
    if is_forest and "test_infer_ms_per_img_CPU" in row.index and pd.notna(row["test_infer_ms_per_img_CPU"]):
        return float(row["test_infer_ms_per_img_CPU"])
    if "gpu_latency_ms" in row.index and pd.notna(row["gpu_latency_ms"]):
        return float(row["gpu_latency_ms"])
    if "cpu_latency_ms" in row.index and pd.notna(row["cpu_latency_ms"]):
        return float(row["cpu_latency_ms"])
    if "test_infer_ms_per_img_CPU" in row.index and pd.notna(row["test_infer_ms_per_img_CPU"]):
        return float(row["test_infer_ms_per_img_CPU"])
    return np.nan

forest_row  = get_row(benchmark_df, "Forest_Ensemble")
expert1_row = get_row(benchmark_df, "Expert_1 (efficientnet_b0)")
expert2_row = get_row(benchmark_df, "Expert_2 (efficientnet_b1)")
expert3_row = get_row(benchmark_df, "Expert_3 (efficientnet_b2)")
student_row = get_row(benchmark_df, "Student_MobileNetV3_Small")

paper_rows = [
    {
        "Model": "Expert 1 (B0)",
        "Params (M)": round(float(expert1_row["params_M"]), 2),
        "GMACs": round(float(expert1_row["FLOPs_GMACs"]), 2),
        "GFLOPs": round(2 * float(expert1_row["FLOPs_GMACs"]), 2),
        "Val Acc (%)": round(float(expert1_row["val_acc"]) * 100, 2),
        "Test Acc (%)": round(float(expert1_row["test_acc"]) * 100, 2),
        "Latency (ms)": round(pick_latency_ms(expert1_row), 2),
    },
    {
        "Model": "Expert 2 (B1)",
        "Params (M)": round(float(expert2_row["params_M"]), 2),
        "GMACs": round(float(expert2_row["FLOPs_GMACs"]), 2),
        "GFLOPs": round(2 * float(expert2_row["FLOPs_GMACs"]), 2),
        "Val Acc (%)": round(float(expert2_row["val_acc"]) * 100, 2),
        "Test Acc (%)": round(float(expert2_row["test_acc"]) * 100, 2),
        "Latency (ms)": round(pick_latency_ms(expert2_row), 2),
    },
    {
        "Model": "Expert 3 (B2)",
        "Params (M)": round(float(expert3_row["params_M"]), 2),
        "GMACs": round(float(expert3_row["FLOPs_GMACs"]), 2),
        "GFLOPs": round(2 * float(expert3_row["FLOPs_GMACs"]), 2),
        "Val Acc (%)": round(float(expert3_row["val_acc"]) * 100, 2),
        "Test Acc (%)": round(float(expert3_row["test_acc"]) * 100, 2),
        "Latency (ms)": round(pick_latency_ms(expert3_row), 2),
    },
    {
        "Model": "Forest Teacher",
        "Params (M)": round(float(forest_row["params_M"]), 2),
        "GMACs": round(float(forest_row["FLOPs_GMACs"]), 2),
        "GFLOPs": round(2 * float(forest_row["FLOPs_GMACs"]), 2),
        "Val Acc (%)": round(float(forest_row["val_acc"]) * 100, 2),
        "Test Acc (%)": round(float(forest_row["test_acc"]) * 100, 2),
        "Latency (ms)": round(pick_latency_ms(forest_row, is_forest=True), 2),
    },
    {
        "Model": "Student (Ours)",
        "Params (M)": round(float(student_row["params_M"]), 2),
        "GMACs": round(float(student_row["FLOPs_GMACs"]), 2),
        "GFLOPs": round(2 * float(student_row["FLOPs_GMACs"]), 2),
        "Val Acc (%)": round(float(student_row["val_acc"]) * 100, 2),
        "Test Acc (%)": round(float(student_row["test_acc"]) * 100, 2),
        "Latency (ms)": round(pick_latency_ms(student_row), 2),
    },
]

paper_df = pd.DataFrame(paper_rows)
paper_df.to_csv(TABLES_DIR / "table_results_paper_style.csv", index=False)
print("Saved:", TABLES_DIR / "table_results_paper_style.csv")
print(paper_df)

# -------------------------
# Paper plots (same filenames used in LaTeX)
# -------------------------
labels_for_plots   = paper_df["Model"].tolist()
val_accs_for_plots = paper_df["Val Acc (%)"].tolist()
test_accs_for_plots = paper_df["Test Acc (%)"].tolist()
params_for_plots   = paper_df["Params (M)"].tolist()
gmacs_for_plots    = paper_df["GMACs"].tolist()
latency_for_plots  = paper_df["Latency (ms)"].tolist()

plot_bar(
    val_accs_for_plots,
    labels_for_plots,
    ylabel="Validation Accuracy (%)",
    title="Validation Accuracy Across Models",
    out_path=PLOTS_DIR / "plot_val_accuracy.png",
)
plot_bar(
    test_accs_for_plots,
    labels_for_plots,
    ylabel="Test Accuracy (%)",
    title="Test Accuracy Across Models",
    out_path=PLOTS_DIR / "plot_test_accuracy.png",
)
plot_bar(
    params_for_plots,
    labels_for_plots,
    ylabel="Params (M)",
    title="Model Size Comparison",
    out_path=PLOTS_DIR / "plot_params.png",
)
plot_dual_metric(
    labels_for_plots,
    gmacs_for_plots,
    latency_for_plots,
    out_path=PLOTS_DIR / "panel_d_gmacs_latency.png",
)

# -------------------------
# Manifest
# -------------------------
manifest = {
    "class_names": class_names,
    "num_classes": len(class_names),
    "forest_weights": forest_weights_np.tolist(),
    "saved_files": {
        "benchmark_csv": str(ARTIFACT_ROOT / "full_benchmark_results.csv"),
        "paper_table_csv": str(TABLES_DIR / "table_results_paper_style.csv"),
        "macro_table_csv": str(TABLES_DIR / "table_macro_metrics.csv"),
        "mcnemar_table_csv": str(TABLES_DIR / "table_mcnemar.csv"),
        "plot_val_accuracy": str(PLOTS_DIR / "plot_val_accuracy.png"),
        "plot_test_accuracy": str(PLOTS_DIR / "plot_test_accuracy.png"),
        "plot_params": str(PLOTS_DIR / "plot_params.png"),
        "panel_d_gmacs_latency": str(PLOTS_DIR / "panel_d_gmacs_latency.png"),
        "confusion_combined_vertical": str(PLOTS_DIR / "confusion_combined_vertical.png"),
    }
}
with open(ARTIFACT_ROOT / "manifest.json", "w") as f:
    json.dump(manifest, f, indent=2)

print("\nDone. All paper artifacts saved under:", ARTIFACT_ROOT)
```
